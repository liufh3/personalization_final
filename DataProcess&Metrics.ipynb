```python
import pandas as pd
import numpy as np
import json
from tqdm import tqdm_notebook
import collections
import time
import bisect
```

```python
def load_json(file_name, save = True):
    lines = open("yelp_dataset/"+str(file_name)+".json").readlines()
    keys = list(json.loads(lines[0]).keys())
    content_dict = collections.defaultdict(list)
    for line in tqdm_notebook(lines):
        blob = json.loads(line)
        for k in keys:
            content_dict[k].append(blob[k])
    df = pd.DataFrame(content_dict)
    if save:
        df.to_csv(file_name+'.csv',index = 0)
    return df
```

```python
for file_name in ['tip','user']:
    load_json(file_name)
```

```python
def mae(true, pred):
    '''
    true as a vector
    pred as a vector
    calculating mean absolute error
    '''
    assert len(true) == len(pred)
    return np.sum(np.abs(true - pred))

def mse(true, pred):
    '''
    true as a vector
    pred as a vector
    calculating mean square error
    '''
    assert len(true) == len(pred)
    return np.sum(np.abs(true - pred) * np.abs(true - pred))

def get_rank(arraylike):
    score = arraylike['rating']
    user = arraylike['user_id']
    return sum_dict[user][int(score + 0.5) - 1]

def get_rank_pred(arraylike):
    score = arraylike['prediction']
    user = arraylike['user_id']
    return sum_dict[user][int(score + 0.5) - 1]

def rank_accuracy(sum_dict, prediction):
    '''
    input prediction dataframe and make user_id as index
    '''
    rank_true = prediction.apply(get_rank,axis = 1)
    rank_pred = prediction.apply(get_rank_pred,axis = 1)
    return np.corrcoef(rank_true,rank_pred)[0][1]  
```
